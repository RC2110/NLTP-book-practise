```python
file = open("miracle_in_the_andes.txt", 'r', encoding='UTF-8')
content = file.read()
```

```python
content.count("Chapter")
```

```python
import re
```

```python
pattern =  re.compile("Chapter [0-9]+")
findings = re.findall(pattern, content)
print(findings)
len(findings)
```

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings =  re.findall(pattern, content)
len(findings)
```

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, content.lower())
len(findings)
```

```python
data={}
for i in findings:
    if i in data.keys():  
        data[i] = data[i] + 1
    else:
        data[i]=1
```

```python
lst=[(values, keys) for (keys, values) in data.items()]
```

```python
sorted(lst, reverse=True)[0:10]
```
